```python
# 名称列表
names = [
    "口腔运送",
    "软腭上抬",
    "舌骨运动",
    "食管上括约肌开放",
    "咽期吞咽启动",
    "咽腔运送",
    "喉前庭关闭",
    "整体吞咽"
]

# 数值列表
values = [
    [0.427323, 0.439129],
    [0.428548, 0.464645],
    [0.437097, 0.479226],
    [0.443, 0.467484],
    [0.434613, 0.439258],
    [0.437419, 0.469387],
    [0.441452, 0.462355]
]

# 组合名称和数值
result = []
for name, area in zip(names, values):
    result.append({
        "name": name,
        "areas": [{"start": area[0], "end": area[1]}]
    })

# 输出结果
import json
print(json.dumps(result, indent=2))
```
